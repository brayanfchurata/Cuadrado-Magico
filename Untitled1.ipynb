```python
import tkinter as tk
import itertools

# Función para verificar si es un cuadrado mágico
def es_cuadrado_magico(matriz):
    suma_magica = 15
    for i in range(3):
        if sum(matriz[i]) != suma_magica or sum(row[i] for row in matriz) != suma_magica:
            return False
    if sum(matriz[i][i] for i in range(3)) != suma_magica or sum(matriz[i][2-i] for i in range(3)) != suma_magica:
        return False
    return True

# Función para llenar la cuadrícula visualmente
def llenar_cuadro(matriz):
    for i in range(3):
        for j in range(3):
            entradas[i][j].config(state='normal')
            entradas[i][j].delete(0, tk.END)
            entradas[i][j].insert(0, str(matriz[i][j]))
            entradas[i][j].config(state='readonly')

# Función para buscar y mostrar iteraciones automáticamente
def buscar_cuadrados_automaticamente():
    nums = list(range(1, 10))
    for perm in itertools.permutations(nums):
        matriz = [list(perm[i:i+3]) for i in range(0, 9, 3)]
        panel_iteraciones.insert(tk.END, f"Probando: {perm}\n")
        ventana.update_idletasks()  # Actualizar la interfaz
        
        if es_cuadrado_magico(matriz):
            llenar_cuadro(matriz)
            mensaje_var.set("¡Cuadrado mágico encontrado!")
            panel_iteraciones.insert(tk.END, f"Cuadrado mágico: {matriz}\n")
            break  # Detener después del primer cuadrado mágico encontrado
    else:
        mensaje_var.set("No se encontró ningún cuadrado mágico.")

# Crear ventana principal
ventana = tk.Tk()
ventana.title("Cuadrado Mágico - Búsqueda Automática")
ventana.geometry("900x500")
ventana.config(bg="#e6f7ff")

# Título
titulo = tk.Label(ventana, text="Cuadrado Mágico 3x3", font=("Helvetica", 24, "bold"), bg="#e6f7ff")
titulo.pack(pady=10)

# Cuadrícula
frame_cuadro = tk.Frame(ventana, bg="#e6f7ff")
frame_cuadro.pack(side=tk.LEFT, padx=20, pady=20)

entradas = [[tk.Entry(frame_cuadro, width=5, font=('Arial', 30), justify='center', state='readonly') for _ in range(3)] for _ in range(3)]
for i in range(3):
    for j in range(3):
        entradas[i][j].grid(row=i, column=j, padx=10, pady=10)

# Panel lateral de iteraciones
frame_iteraciones = tk.Frame(ventana, bg="#ffffff", relief=tk.SUNKEN, borderwidth=3)
frame_iteraciones.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True, padx=10)

panel_titulo = tk.Label(frame_iteraciones, text="Iteraciones", font=("Helvetica", 16, "bold"), bg="#ffffff")
panel_titulo.pack(pady=5)

panel_iteraciones = tk.Text(frame_iteraciones, font=("Courier", 12), bg="#f7f7f7", height=20, width=40)
panel_iteraciones.pack(pady=10, padx=5)

# Mensaje dinámico
mensaje_var = tk.StringVar()
mensaje_var.set("Iniciando búsqueda automática de cuadrados mágicos.")
mensaje_label = tk.Label(ventana, textvariable=mensaje_var, font=("Helvetica", 14), fg="#007acc", bg="#e6f7ff")
mensaje_label.pack(pady=10)

# Ejecutar búsqueda automáticamente al iniciar la ventana
ventana.after(1000, buscar_cuadrados_automaticamente)

# Ejecutar la interfaz
ventana.mainloop()
```
